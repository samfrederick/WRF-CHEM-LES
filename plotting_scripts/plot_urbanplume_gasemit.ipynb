```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas_emit = pd.read_csv('urban_plume_gas_emit.csv')
```

```python
species = ['SO2', 'NO2', 'NO', 'NH3', 'CO', 'ALD2', 'HCHO', 'ETH', 'OLEI', 
           'OLET', 'TOL', 'XYL', 'AONE', 'PAR', 'ISOP', 'CH3OH', 'ANOL']
```

```python
# show timeseries for the first five species
fig, ax = plt.subplots(1,1, figsize=(12, 4))
for param in species[:5]:
    ax.plot(gas_emit.time_s/3600, gas_emit[param], label=param)
ax.set_yscale('log')
plt.legend()
ax.set_xlabel('Time (hrs)')
ax.set_ylabel('Emission Rate (mol$\cdot $m$^{-2}\cdot$ s$^{-1}$)')
```

```python
#TODO transform emission rates from mol m^-2 s^-1 to mol km^-2 hr^-1 by multipying by 3.6E9

# show timeseries for the first five species
fig, ax = plt.subplots(1,1, figsize=(12, 4))
for param in species[:5]:
    ax.plot(gas_emit.time_s/3600, 3.6E9*gas_emit[param], label=param)
ax.set_yscale('log')
plt.legend()
ax.set_xlabel('Time (hrs)')
ax.set_ylabel('Emission Rate (mol$\cdot $km$^{-2}\cdot$ hr$^{-1}$)')
```
